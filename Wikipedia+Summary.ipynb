# extracting text from soup

# Pre Processing Data

```python
#text = re.sub(r'[^\x00-\x7F]+',' ', text)
text = re.sub(r"\[[0-9]\]*",' ',text) # removing digits from square brackets like [23]
text = re.sub(r'[0-9]+\]',' ',text)
text = re.sub(r'\s+',' ',text) # remove digit from single square bracket like 9]
#text = re.sub(r'\W',' ',text)
text
```

```python
cleantext = text.lower()
cleantext = re.sub(r'\W',' ',cleantext)
cleantext = re.sub(r'[0-9]+',' ',cleantext)
cleantext = re.sub(r'\s+',' ',cleantext)
cleantext = re.sub(r'\s+[a-z]\s+',' ',cleantext)
cleantext
```

```python
text
```

```python
import pandas as pd
import numpy as np
import heapq
from nltk.corpus import stopwords
```

```python
sentences = nltk.sent_tokenize(text)
stop_words = set(stopwords.words('english'))
```

```python
sentences
```

# Word Count Manually ie Normal Histogram

```python
word2count = {}
for word in nltk.word_tokenize(cleantext):
    if word not in stop_words:
        if word not in word2count.keys():
            word2count[word] = 1
        else :
            word2count[word]+= 1
freq_words = heapq.nlargest(len(word2count),word2count,key = word2count.get)  # we can replace len with some fixed numbers          
```

# Calculating weight of each word ie Weighted Histogram

```python
for key in word2count.keys():
    word2count[key] = word2count[key] / max(word2count.values())
```

```python
word2count
```

# Calculating Scores For Each Sentences

```python
sent2score ={}
for sentence in sentences:
    for word in nltk.word_tokenize(sentence.lower()):
        if word in word2count.keys():
            if (len(sentence.split(' '))) < 30:
                if sentence not in sent2score.keys():
                    sent2score[sentence] = word2count[word]
                else :
                    sent2score[sentence] += word2count[word]
```

# Summary

```python
bestSentences = heapq.nlargest(5,sent2score,key = sent2score.get)
```

```python
print('==================================================== Summary ===============================================')
for sentence in bestSentences:
    print(sentence)
```
